```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
file_path = 'Dataset.csv'  # The file is in the same directory as your Jupyter Notebook
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: File not found at path: {file_path}.  Make sure 'Dataset.csv' is in the same directory as your notebook, or provide the correct path.")
    exit()  # Stop execution if the file isn't found


# Filter for moderate to high adaptivity
df_adaptable = df[df['Adaptivity Level'].isin(['Moderate', 'High'])].copy()

# Calculate the proportion of IT Students
it_student_counts = df_adaptable['IT Student'].value_counts(normalize=True) * 100

# Calculate the proportion of Internet Types
internet_type_counts = df_adaptable['Internet Type'].value_counts(normalize=True) * 100

# Calculate the proportion of devices used
device_counts = df_adaptable['Device'].value_counts(normalize=True) * 100

# Define colors
colors = ['#ff9999', '#66b3ff', '#99ff99', '#c2c2f0']

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(24, 8))  # Increased figure size for better label display

# Add overall explanation text ABOVE the plots
fig.text(0.5, 0.99,
         "These pie charts show the distribution of key factors among students with moderate to high adaptability in online learning.\n"
         "Each chart represents a different factor (IT Student status, Internet Type, and Device). The percentages indicate the proportion\n"
         "of students in this adaptable group who possess that characteristic.  For example, a larger slice for 'Wifi' suggests a strong\n"
         "association between reliable Wifi and adaptability.  Consider these factors as potential indicators, not definitive causes, of success.\n",
         ha="center", fontsize=18, wrap=True)


fig.suptitle('Key Indicators of Moderate to High Adaptability in Online Learning', fontsize=20)  # Increased title font size

# Plot pie chart for IT Student
labels_it = [f"{label} ({value:.1f}%)" for label, value in it_student_counts.items()]
axes[0].pie(it_student_counts, labels=labels_it, autopct='', shadow=True, startangle=140, colors=colors[:len(it_student_counts)], textprops={'fontsize': 12})  # Increased label font size
axes[0].set_title('IT Student\n(Percentage of Moderate/High Adaptability Students)', fontsize=16)  # Increased title font size
axes[0].axis('equal')

# Plot pie chart for Internet Type
labels_internet = [f"{label} ({value:.1f}%)" for label, value in internet_type_counts.items()]
axes[1].pie(internet_type_counts, labels=labels_internet, autopct='', shadow=True, startangle=140, colors=colors[:len(internet_type_counts)], textprops={'fontsize': 12})  # Increased label font size
axes[1].set_title('Internet Type\n(Percentage of Moderate/High Adaptability Students)', fontsize=16)  # Increased title font size
axes[1].axis('equal')

# Plot pie chart for Device
labels_device = [f"{label} ({value:.1f}%)" for label, value in device_counts.items()]
axes[2].pie(device_counts, labels=labels_device, autopct='', shadow=True, startangle=140, colors=colors[:len(device_counts)], textprops={'fontsize': 12})  # Increased label font size
axes[2].set_title('Device\n(Percentage of Moderate/High Adaptability Students)', fontsize=16)  # Increased title font size
axes[2].axis('equal')

# Ensure the circle's proportion
for ax in axes:
    ax.axis('equal')

# Adjust layout to prevent labels from overlapping
plt.tight_layout(rect=[0.03, 0.03, 0.97, 0.92])

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
file_path = 'Dataset.csv'  # The file is in the same directory as your Jupyter Notebook
df = pd.read_csv(file_path)

# Filter for moderate to high adaptivity
df_adaptable = df[df['Adaptivity Level'].isin(['Moderate', 'High'])].copy()

# --- Data Preparation for Bar Charts ---

# 1. IT Student:
it_student_counts = df_adaptable['IT Student'].value_counts()

# 2. Internet Type:
internet_type_counts = df_adaptable['Internet Type'].value_counts()

# 3. Device:
device_counts = df_adaptable['Device'].value_counts()


# --- Plotting Bar Charts ---

fig, axes = plt.subplots(1, 3, figsize=(24, 8))
fig.suptitle('Key Indicators of Moderate to High Adaptability in Online Learning (Bar Chart)', fontsize=20)

# Add overall explanation text ABOVE the charts
fig.text(0.5, 0.99,  # Adjusted position to place text at the top
         "NOTE:\n"
         "These bar charts show the distribution of key factors among students with moderate to high adaptability in online learning.\n"
         "Each chart represents a different factor(IT student status, internet type, and device). The height of each bar indicates the number\n"
         "of students in this adaptable group who possess that characteristic.  For example, a taller bar for 'Wifi' suggests a strong\n"
         "association between reliable Wifi and adaptability.  Consider these factors as potential indicators, not definitive causes, of success.\n",
         ha="center", fontsize=18, wrap=True)


# 1. IT Student Bar Chart
axes[0].bar(it_student_counts.index, it_student_counts.values, color=['skyblue', 'lightcoral'])
axes[0].set_title('IT Student\n(Counts of Moderate/High Adaptability Students)', fontsize=16)
axes[0].set_xlabel('IT Student', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].tick_params(axis='x', labelsize=10)
axes[0].tick_params(axis='y', labelsize=10)


# 2. Internet Type Bar Chart
axes[1].bar(internet_type_counts.index, internet_type_counts.values, color=['lightgreen', 'lightsalmon', 'lightblue'])
axes[1].set_title('Internet Type\n(Counts of Moderate/High Adaptability Students)', fontsize=16)
axes[1].set_xlabel('Internet Type', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].tick_params(axis='x', labelsize=10, rotation=45)  # Rotate x-axis labels for readability
axes[1].tick_params(axis='y', labelsize=10)


# 3. Device Bar Chart
axes[2].bar(device_counts.index, device_counts.values, color=['lightseagreen', 'plum', 'gold'])
axes[2].set_title('Device\n(Counts of Moderate/High Adaptability Students)', fontsize=16)
axes[2].set_xlabel('Device', fontsize=12)
axes[2].set_ylabel('Count', fontsize=12)
axes[2].tick_params(axis='x', labelsize=10)
axes[2].tick_params(axis='y', labelsize=10)



plt.tight_layout(rect=[0, 0.03, 1, 0.92])  # Adjust layout to make room for the title and explanation

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
file_path = 'Dataset.csv'  # The file is in the same directory as your Jupyter Notebook
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"Error: File not found at path: {file_path}.  Make sure 'Dataset.csv' is in the same directory as your notebook, or provide the correct path.")
    exit()  # Stop execution if the file isn't found


# --- Data Preparation for Histograms ---

# Let's create histograms for 'Age' and 'Class Duration'
# You might need to adjust bin sizes depending on your data

# Convert 'Age' and 'Class Duration' to appropriate numeric types if needed
df['Age'] = df['Age'].astype(str)  # Ensure it's treated as a string
df['Age_Lower'] = df['Age'].str.split('-').str[0].astype(int) # take the lower bound of age range
df['Class Duration'] = df['Class Duration'].astype(str)
df['Class Duration_Lower'] = df['Class Duration'].str.split('-').str[0].astype(int) # take the lower bound of class duration range

# --- Plotting Histograms ---

fig, axes = plt.subplots(1, 2, figsize=(24, 8))  # Adjust figure size as needed

# Add overall explanation text ABOVE the plots
fig.text(0.5, 0.99,  # Adjusted y position
         "These histograms show the distribution of student age and class duration in the dataset.\n"
         "The x-axis of each histogram represents the age (lower bound of the range) or class duration (lower bound of the range in hours),\n"
         "and the y-axis represents the frequency, or number of students falling into each bin (range of ages or class durations).\n"
         "The shape of the histogram provides insights into the typical age and class duration patterns within the student population.\n"
         "Consider these distributions when analyzing factors related to student adaptability in online learning.\n",
         ha="center", fontsize=18, wrap=True)


fig.suptitle('Distribution of Age and Class Duration', fontsize=20)

# 1. Age Histogram
axes[0].hist(df['Age_Lower'], bins=10, color='skyblue', edgecolor='black')  # Adjust bins as needed
axes[0].set_title('Distribution of Student Age', fontsize=16)
axes[0].set_xlabel('Age (Lower Bound of Range)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].tick_params(axis='x', labelsize=10)
axes[0].tick_params(axis='y', labelsize=10)

# 2. Class Duration Histogram
axes[1].hist(df['Class Duration_Lower'], bins=5, color='lightgreen', edgecolor='black')  # Adjust bins as needed
axes[1].set_title('Distribution of Class Duration', fontsize=16)
axes[1].set_xlabel('Class Duration (Hours, Lower Bound of Range)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].tick_params(axis='x', labelsize=10)
axes[1].tick_params(axis='y', labelsize=10)


plt.tight_layout(rect=[0.03, 0.03, 0.97, 0.92]) # Adjust layout for title and explanation

plt.show()
```
